```python
click_month = driver.find_element_by_xpath('//*[@id="pageContent"]/div[2]/table/tbody/tr[1]/td/span/div/strong/a[235]').click()
```

```python
#select the assigned month
date_url = []
for d in driver.find_elements_by_xpath('//td[@align="center"]//a'):
    date_url.append(d.get_attribute('href'))
date_url
```

```python
Click_date = driver.find_element_by_xpath('//*[@id="calender"]/tbody/tr[2]/td[3]').click()
```

```python
News_Article = []

for url1 in date_url:
    driver.get(url1)
    
    for n in driver.find_elements_by_xpath('//td[@class="contentbox5"]/ul/li/a'):
        News_Article.append(n.get_attribute('href'))
News_Article
```

```python
Date = []
Auuthor = []
Vertical = []
Headline = []
Description = []

for url2 in News_Article:
    driver.get(url2)
    
    for c in driver.find_elements_by_xpath('//div[@class="dtc vam artByline"]//time'):
        Date.append(c.text)
        
    for a in driver.find_elements_by_xpath('//span[@class="ag"]'):
        Auuthor.append(a.text)

    for v in driver.find_elements_by_xpath('//a[@itemprop="item"]'):
        Vertical.append(v.text)
        
    for h in driver.find_elements_by_xpath('//h1[@class="artTitle font_faus"]'):
        Headline.append(h.text)
        
    for i in driver.find_elements_by_xpath('//div[@class="artText medium"]'):
        Description.append(i.text)
    

News = pd.DataFrame({})
News['Date'] = Date
News['Auuthor']= Author
News['Vertical']=Vertical or Topic
News['Headline']= Headlines
News['Description']= Description

News
```
